Word Cloud is a data visualization technique used for representing text data. The size of each word indicates its frequency or importance.<br>
Word clouds are widely used for analyzing data from social network websites as significant textual data points can be highlighted using a word cloud. 

For generating word cloud in Python, modules needed are â€“ matplotlib, pandas and wordcloud:

### Example

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd

df = pd.read_csv(r"Files\YouTube-Spam-Collection-v1\Youtube04-Eminem.csv", encoding ="latin-1") 
# Reads 'Youtube04-Eminem.csv' file

comment_words = ''
stopwords = set(STOPWORDS)

for eachValue in df.CONTENT:
    
    value = str(eachValue).lower()
    # Transform 'eachValue' into a string object
    
    comment_words += value + " "
    
wordcloud = WordCloud(width = 800, height = 800,
                      background_color ='white',
                      stopwords = stopwords,
                      min_font_size = 10).generate(comment_words)

# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)
```

## Sentiment from a DataFrame

The data we will be using has be a 'sentiment' column, showing whether the tweet is positive or negative.<br>
We can use this data to build word clouds for positive tweets and negative tweets:

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd

cleanData = pd.read_csv(r"Files\cleanedData.csv")

print(cleanData.head())

df_senti_list = [x[1] for x in cleanData.groupby('sentiment', sort=False)]
# Creates separate lists depending on the sentiment of the row.
    # In this case, 3 lists are made - Positive, Negative, Neutral
    
# .describe() methods can't be printed all at once.

print("\nDataFrame for Neutral tweets:")
df_senti_list[0].describe()
# [0] = DataFrame for Neutral tweets.

print("\nDataFrame for Negative tweets:")
df_senti_list[1].describe()
# [1] = DataFrame for Negative tweets.

print("\nDataFrame for Positive tweets:")
df_senti_list[2].describe()
# [2] = DataFrame for Positive tweets.
```

Now that we have tweets separated out into positive, negative and neutral sentiments, we can create Word Clouds for each individual sentiment:

```python
""" Carrying on from above code ... """

# NEUTRAL WordCloud

comment_words = ''
stopwords = set(STOPWORDS)

for eachValue in df_senti_list[0].text:
    
    eachValue = eachValue.replace("[","").replace("]","").replace("'","")
    # 'eachValue' is a String object but [] like a List object.

    valuesList = eachValue.split(",")
    
    for value in valuesList:
        
        comment_words += " " + value.strip().lower()

wordcloud_Neutral = WordCloud(width = 800, height = 800,
                              background_color ='white',
                              stopwords = stopwords,
                              min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud_Neutral) 
plt.axis("off") 
plt.tight_layout(pad = 0)
```

```python
""" Carrying on from above code ... """

# NEGATIVE WordCloud

comment_words = ''
stopwords = set(STOPWORDS)

for eachValue in df_senti_list[1].text:
    
    eachValue = eachValue.replace("[","").replace("]","").replace("'","")
    # 'eachValue' is a String object but [] like a List object.

    valuesList = eachValue.split(",")
    
    for value in valuesList:
        
        comment_words += " " + value.strip().lower()

wordcloud_Negative = WordCloud(width = 800, height = 800,
                              background_color ='white',
                              stopwords = stopwords,
                              min_font_size = 10).generate(comment_words)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud_Negative)
plt.axis("off") 
plt.tight_layout(pad = 0)
```
